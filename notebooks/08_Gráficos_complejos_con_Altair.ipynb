# Gráficos complejos y funciones avanzadas de `Altair`

En esta lección vamos a ver:
1. Agregaciones con Altair (no en pandas)
2. Cómo crear facetas y concatenar gráficos
3. Capas de gráficos 

```
import altair as alt
import seaborn as sns
```

```
pinguinos = sns.utils.load_dataset("penguins")
```

```
pinguinos.head()
```

### Agregaciones

```
alt.Chart(pinguinos).mark_bar().encode(
    x = alt.X("mean(body_mass_g):Q"),
    y = alt.Y("species:N")
)
```

```
alt.Chart(pinguinos).mark_bar().encode(
    x = alt.X("count()"),
    y = alt.Y("island:N")
)
```

### Facetas

```
alt.Chart(pinguinos).mark_bar().encode(
    x = "count()",
    y = alt.Y("island:N"),
    row = alt.Row("species:N")
)
```

```
alt.Chart(pinguinos).mark_circle().encode(
    x = alt.X("bill_length_mm:Q"),
    y = alt.Y("bill_depth_mm:Q"),
    row = alt.Row("island:N"),
    column = alt.Column("species:N")
).properties(
    width = 100,
    height= 100,
    title = "Pinguinos yay"
)
```

### Concatenar gráficos

```
barras = alt.Chart(pinguinos).mark_bar().encode(
    x = "species:N",
    y = alt.Y("mean(body_mass_g)")
)

dispersion = alt.Chart(pinguinos).mark_circle().encode(
    x = alt.X("bill_length_mm"),
    y = alt.Y("bill_depth_mm"),
    color = alt.Color("species:N")
)

par1 = alt.hconcat(barras, dispersion)

alt.vconcat(par1, par1)
```

### Capas

```
dispersion_base = alt.Chart(pinguinos).mark_circle(fillOpacity = 0.50).encode(
    x = alt.X("bill_length_mm:Q", title = "Largo del pico (mm)"),
    y = alt.Y("bill_depth_mm:Q", title = "Profundidad del pico (mm)"),
    color = alt.Color("species:N")
)

dispersion_texto = alt.Chart(pinguinos).mark_text(size = 16).encode(
    x = alt.X("mean(bill_length_mm):Q", ),
    y = alt.Y("mean(bill_depth_mm):Q"),
    text = alt.Text("species:N")
)

dispersion_base + dispersion_texto
```
